# Time-series inversion example

This notebook shows how to extract a cumulative displacement time-series from a network of pairwise displacement measurements through inversion using a synthetic signal as example. Initially we will work with a network of perfect measurements and then experiment with the addition of different noise types.

```python
# load neccessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
from itertools import combinations
import random

#import functions for time-series inversion from relevant python script
import timeseries_inversion as inv
```

## Noise-free example

As a first example, we will create a synthetic displacement time-series and sample this signal for different date pairs. These are then used in the inversion process to reconstruct the original signal. 

```python
#create an artificial displacement original_signal
start_date = datetime.strptime('01/01/2019', '%d/%m/%Y')
end_date = datetime.strptime('31/12/2023', '%d/%m/%Y')

dr = pd.date_range(start=start_date, end=end_date, freq='D')


#create synthetic cumulative displacement original_signal
#here: sigmoid + linear function
x = np.arange(0, len(dr), 1)
y = 0.01 * x + 20 / (1 + np.exp(-0.02 * (np.arange(len(dr)) - len(dr)/2)))
original_signal = pd.DataFrame({'date': dr.date, 'disp': y})

plt.figure()
plt.plot(original_signal.date, original_signal.disp)
plt.xlabel('Date')
plt.ylabel('Displacement')
```

```python
# define dates on which to sample the displacement signal (usually dates of satellite acquisitions)

# option 1: regular sampling every 3 months
sample_dates = pd.date_range(start=start_date, end=end_date, freq='3MS').date

#option 2: irregular sampling by drawing a 30 random dates from the date range defined earlier 
random.seed(123)
sample_dates = random.sample(list(dr.date), 30)
sample_dates.append(datetime.date(start_date)) #make sure start date is included
sample_dates.sort()

#create image pairs from sample dates
date_combinations = list(combinations(sample_dates, 2))

network = pd.DataFrame(date_combinations, columns=['date0', 'date1'])

#find displacement for each datepair according to the synthetic signal
network['disp'] = network.apply(lambda row: inv.find_dsoll(row, original_signal), axis = 1)

#run inversion
timeseries = inv.run_inversion(network)
```

```python
#plot results
inv.plot_results(timeseries, original_signal)
```

## Addition of noise

Unlike in this first example, displacement measurements derived from image cross-correlation are usually not free of noise. In the next step, we will simulate the effect of different noise types on the inverted time-series.

```python
# Gaussian noise

np.random.seed(123)
noise = np.random.normal(0,5,len(network))

# add the Gaussian noise to the displacement measured for each date pair
network2 = network.copy()
network2['disp'] = network.disp + noise

#rerun the inversion
timeseries = inv.run_inversion(network2)
inv.plot_results(timeseries, original_signal)
```

```python
# Fixed image Gaussian noise

network3 = network.copy()

np.random.seed(123)
#get fixed noise value for every image
noise = np.random.normal(0,5,len(sample_dates))

#merge noise values with network df
m = pd.DataFrame({"date": sample_dates, "noise":noise})
merge = network3.merge(m, left_on = "date0", right_on = "date")
merge = merge.merge(m, left_on = "date1", right_on = "date")

#noise value of a correlation will be the sum of noise associated with reference and secondary image
noise = merge.noise_x + merge.noise_y 

# add noise to the displacement measured for each date pair
network3['disp'] = network.disp + noise

#rerun the inversion
timeseries = inv.run_inversion(network3)
inv.plot_results(timeseries, original_signal)
```

```python
# Seasonal noise

network4 = network.copy()


#define seasonal signal as sine wave with one year wavelength
dt = (network.date1 - network.date0).dt.days
dt_rad = (dt / (365.25/2)) * ( np.pi)

#set wave amplitude to 15
noise = np.sin(dt_rad) * 15


plt.figure()
plt.scatter(dt,noise)
plt.title("Added noise")
plt.xlabel("Temporal baseline")
plt.ylabel("Noise")
```

```python
# add noise to the displacement measured for each date pair
network4['disp'] = network.disp + noise

#rerun the inversion
timeseries = inv.run_inversion(network4)
inv.plot_results(timeseries, original_signal)
```
